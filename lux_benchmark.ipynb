# Comparison on multiple datasets

```python
from lux.lux import LUX
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np
import pandas as pd
import os
import shap
import sys
sys.path.append('./pyuid3')

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from pyuid3.data import Data
import pickle
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.neighbors import NearestNeighbors
import graphviz

from sklearn import tree
from sklearn.tree import  export_graphviz
from graphviz import Source
from IPython.display import display
from IPython.display import SVG
```

```python
def minmaxdf(df):
    maxv = df.max()
    return df/maxv
```

```python
nac_data  = []
rs=42
fraction=0.1
directory = './data/numerical_only/balanced'
up=True
depth = 2

for num_and_cat in  os.listdir(directory):
    #(data, y, categorical_indicator) = pickle.load(open(f'{directory}/{num_and_cat}','rb'))
    (data, y) = pickle.load(open(f'{directory}/{num_and_cat}','rb'))
    features = [f'f_{i}' for i in range(0,data.shape[1])]
    target = 'class'
    df = pd.DataFrame(data,columns=features,dtype=float)
    df[target] = y
    
    ## train xgboost, 
    train, test = train_test_split(df, random_state=rs)
    clf = xgb.XGBClassifier()#RandomForestClassifier(random_state=rs,n_jobs=-1)
    clf.fit(train[features],train[target])
    model_score = clf.score(test[features],test[target])
    print(f'Accuracy  model: {model_score}')
    
    
    explain_instance = train[features].sample(1).values#.astype('f8')
    
    n_neighbors=int(len(train)*0.01)
    nn = NearestNeighbors(n_neighbors=max(1,int(n_neighbors)))
    nn.fit(df[features])
    _,ids_c = nn.kneighbors(np.array(explain_instance).reshape(1,-1))
    dfneigh= df.iloc[ids_c.ravel()]
    
    
    
    data = Data.parse_dataframe(dfneigh[features+[target]])
    
    clf_dtree = DecisionTreeClassifier(random_state=rs,max_depth=depth,criterion='entropy')
    clf_dtree.fit(dfneigh[features],clf.predict(dfneigh[features]))
    
    
    graph = Source(tree.export_graphviz(clf_dtree, out_file=None
       , feature_names=features, class_names=['0', '1'] 
       , filled = True,max_depth=depth))
    display(SVG(graph.pipe(format='svg')))
    
    dectree_score = clf_dtree.score(test[features],clf.predict(test[features]))
    print(f'Fidelity  dectree: {dectree_score}')
    
    #claculate shap, 
    explainer = shap.TreeExplainer(clf, train[features])
    shap_values = explainer.shap_values(train[features],check_additivity=False)
    

    
    #explain xgboost without proba, 
    lux = LUX(predict_proba = lambda x: np.round(clf.predict_proba(x)), neighborhood_size=int(len(train)*fraction),max_depth=depth,  node_size_limit = 2, grow_confidence_threshold = 0 )
    lux.fit(train[features], train[target], instance_to_explain=explain_instance,use_parity=up)
    
    #calculate decision tree score over this dataset
    lux.uid3.tree.save_dot('tree.dot')
    gvz=graphviz.Source.from_file('tree.dot')
    display(gvz)

    
    predictions = [int(x.get_name()) for x in lux.uid3.predict(data.instances)]
    ac_score_pure = accuracy_score(predictions, clf.predict(dfneigh[features]))
    print(f'Fidelity  no proba: {ac_score_pure}')

    
    #with proba, 
    lux = LUX(predict_proba = clf.predict_proba, neighborhood_size=int(len(train)*fraction),max_depth=depth,  node_size_limit = 2, grow_confidence_threshold = 0 )
    lux.fit(train[features], train[target], instance_to_explain=explain_instance,use_parity=up)
    
    lux.uid3.tree.save_dot('tree.dot')
    gvz=graphviz.Source.from_file('tree.dot')
    display(gvz)
    
    predictions = [int(x.get_name()) for x in lux.uid3.predict(data.instances)]
    ac_score_proba = accuracy_score(predictions, clf.predict(dfneigh[features]))
    print(f'Fidelity  with proba: {ac_score_proba}')

    
    #with shap
    # sum for all classess if more than binary classification problem
    if len(shap_values.shape) > 2:
        shap_values=sum([abs(shap) for shap in shap_values])
    else:
        shap_values=abs(shap_values)
    scaled_shap =minmaxdf(shap_values)
    lux = LUX(predict_proba = clf.predict_proba, neighborhood_size=int(len(train)*fraction),max_depth=depth,  node_size_limit = 2, grow_confidence_threshold = 0 )
    lux.fit(train[features], train[target], X_importances = scaled_shap,  instance_to_explain=explain_instance,discount_importance=True,use_parity=up)
    
    lux.uid3.tree.save_dot('tree.dot')
    gvz=graphviz.Source.from_file('tree.dot')
    display(gvz)
    
    predictions = [int(x.get_name()) for x in lux.uid3.predict(data.instances)]
    ac_score_shap = accuracy_score(predictions, clf.predict(dfneigh[features]))
    print(f'Fidelity  with shap: {ac_score_shap}')

    nac_data.append([num_and_cat,ac_score_pure,ac_score_proba,ac_score_shap, model_score, dectree_score])
    
    ## save results
    
    
```

```python
results = pd.DataFrame(nac_data,columns=['num_and_cat','ac_score_pure','ac_score_proba','ac_score_shap', 'model_score', 'dectree_score'])
results.plot()
                       
```

# Wine example

```python
from lux.lux import LUX
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
# import some data to play with
wine = datasets.load_wine()
```

```python
features = wine['feature_names']
target = 'calss'
rs=42

#create daatframe with columns names as strings (LUX accepts only DataFrames withj string columns names)
df_wine = pd.DataFrame(wine.data,columns=features)
df_wine[target] = wine.target

#train classifier
train, test = train_test_split(df_wine, random_state=rs)
clf = RandomForestClassifier(random_state=42)#svm.SVC(probability=True, random_state=rs)
clf.fit(train[features],train[target])
clf.score(test[features],test[target])
```

## Without predictor uncertainty

```python
#pick some instance from datasetr
iris_instance = train[features].sample(1, random_state=42).values
iris_instance

#train lux on neighbourhood equal 30% instances
lux = LUX(predict_proba = lambda x: np.round(clf.predict_proba(x)), neighborhood_size=int(len(train)*fraction),max_depth=2,  node_size_limit = 1, grow_confidence_threshold = 0 )
lux.fit(train[features], train[target], instance_to_explain=iris_instance,class_names=[0,1,2])

#see the justification of the instance being classified for a given class
lux.justify(np.array(iris_instance))
```

```python
iris_instance
```

```python
#calculate decision tree score over this dataset
import sys
sys.path.append('./pyuid3')

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from pyuid3.data import Data
data = Data.parse_dataframe(test[features+[target]])
    
predictions = [int(x.get_name()) for x in lux.uid3.predict(data.instances)]
print(f'Accuracy : {accuracy_score(predictions, test[target])}')
```

```python
#prind the uncertain local tree for the given instance
import graphviz
lux.uid3.tree.save_dot('tree.dot')
graphviz.Source.from_file('tree.dot')
```

## Without SHAP

```python
#pick some instance from datasetr
iris_instance = train[features].sample(1, random_state=42).values
iris_instance

#train lux on neighbourhood equal 20 instances
lux = LUX(predict_proba = clf.predict_proba, neighborhood_size=int(len(train)*fraction),max_depth=2,  node_size_limit = 1, grow_confidence_threshold = 0 )
lux.fit(train[features], train[target], instance_to_explain=iris_instance,class_names=[0,1,2])

#see the justification of the instance being classified for a given class
lux.justify(np.array(iris_instance))
```
